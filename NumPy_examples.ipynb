```python
import numpy as np

t = [1, 3, 5, 6]

print(t[1])
print(type(t[1]))

A = np.array([1, 3, 5, 6])
print(type(A))

print(A[1])
print(type(A[1]))
B = np.array([5, 6, 1, 2])
print(np.shape(A))
print(np.shape(A.T))  # transposition
```

```python
C = np.vstack([A, B])
print(C)
print(np.shape(C))
```
