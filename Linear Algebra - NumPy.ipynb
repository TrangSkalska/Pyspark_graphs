```python
# 2. Generate an array of length 3n filled with the cyclic pattern 1, 2, 3.
```

```python
def cyclic_123(n):
    base = np.array([1, 2, 3])
    return np.tile(base, n)

print(cyclic_123(4))  
```

```python
# 3. Create a 10×10 matrix of zeros and then "frame" it with a border of ones.
```

```python
M = np.zeros((10, 10), dtype=int)
M[0, :] = 1
M[-1, :] = 1
M[:, 0] = 1
M[:, -1] = 1
print(M)
```

```python
# 4. Create a random 3×5 array using the np.random.rand(3, 5) function and compute: the sum of all the entries, the sum of the rows and the sum of the columns. (many Numpy functions have an optional axis= argument!)
```

```python
X = np.random.rand(3, 5)

total_sum = X.sum()
row_sums = X.sum(axis=1)
col_sums = X.sum(axis=0)

print(total_sum)
print(row_sums)
print(col_sums)
```

```python
# 5. Given the following arrays representing logical values (0 = False, 1 = True) compute the logical AND and logical OR operations for every pair of values of the two arrays: [1, 1, 0, 0] [1, 0, 1, 0] Hint: you may need to set the data type (dtype) of the array's elements to 'bool'
```

```python
a = np.array([1, 1, 0, 0], dtype=bool)
b = np.array([1, 0, 1, 0], dtype=bool)

and_result = np.logical_and(a, b)
or_result  = np.logical_or(a, b)

print(and_result)
print(or_result)
```

```python
# 9. Write a function which creates an n×n matrix with (i,j)-entry equal to i+j.
```

```python
def ij_sum_matrix(n):
    i = np.arange(n).reshape(-1, 1)
    j = np.arange(n).reshape(1, -1)
    return i + j

print(ij_sum_matrix(4))
```

```python
# 10. Write a function which creates an n×n matrix with rows having subsequent values multiplied by the row's number. For example for n = 4:
```

```python
def row_scaled_matrix(n):
    base = np.arange(n)         
    rows = np.arange(1, n+1).reshape(-1, 1)  
    return rows * base

print(row_scaled_matrix(4))
```
